```python
from wordcloud import WordCloud
import imageio
import matplotlib.pyplot as plt
import jieba
```

```python
def read_deal_text():
    with open("BV1EP4y1j7kV.txt",encoding='utf-8') as f:    #读取我们的待处理本文
        txt=f.read()
    re_move=["，","。",'\n','\xa0'] #无效数据'""
    #去除无效数据
    for i in re_move:
        txt=txt.replace(i," ") 
    word=jieba.lcut(txt) 
   
    with open("txt_save.txt",'w',encoding='utf-8') as file:
        for i in word:    
            file.write(str(i)+' ') 

```

```python

def img_grearte():
    mask=imageio.imread("a.jpg")
    with open("txt_save.txt",encoding='utf-8') as file:
        txt=file.read()
    word=WordCloud(background_color="white",\
                    width=800,\
                   height=800,
                   font_path='SimHei.ttf',
                   mask=mask,
                   ).generate(txt)
    word.to_file('test.png')
    print("词云图片已保存")
    
    plt.imshow(word)    #使用plt库显示图片
    plt.axis("off")
```

```python

read_deal_text()
img_grearte()
```
